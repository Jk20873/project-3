# Column Checking

```python
# Plot the ethnicity

plt.figure(figsize=(15,8))
ax = sns.countplot(x='Officer-defined ethnicity', data=df)
plt.title('ethnicty define by police officers')
for p in ax.patches:
    x = p.get_bbox().get_points()[:,0]
    
    y = p.get_bbox().get_points()[1,1]
    
    ax.annotate('{:.2g}%'.format(100.*y/len(df)), (x.mean(), y), ha='center', va='bottom')
plt.show()
```

# Observation 
In officer defined ethnicity there was 75% individual was white ,8.9% is Black of the total cases and Asians were also 8.9% ,


```python
# Plot the Outcomes of police reserch
plt.figure(figsize=(15, 8))
ax = sns.countplot(y='Outcome', data=df,
                   order = df.Outcome.value_counts().iloc[:20].index)
plt.title('Outcomes of police actions')
plt.show()
```

# Observation
we can clearly see in above observation that white people arrest more than others but in outcome plot we observe that police takes no action against them only nearly 20% is aressted

```python
# Plot arrest by gender
plt.figure(figsize=(15, 8))

ax = sns.countplot(x = "Gender", data = df,
                   order = df.Gender.value_counts().index)
for p in ax.patches:
    x = p.get_bbox().get_points()[:,0]
    y = p.get_bbox().get_points()[1,1]
    ax.annotate('{:.2g}%'.format(100.*y/len(df)), (x.mean(), y), ha='center', va='bottom')
plt.title('Police Killings by Gender')
plt.show()
```

# Observations

Most of the individuals arrest by police were male
This is expected because females commit less crimes comparative to males
